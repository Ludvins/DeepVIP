```python
import os
import glob
import pandas as pd
os.chdir(".")
```

```python
res1 = [i for i in glob.glob('DVIP_UCI/*')]
df = []
for f in res1:
    try: 
        df.append(pd.read_csv(f))
    except:
        continue
df = pd.concat(df)
df = df.loc[:, (df != df.iloc[0]).any()] 
```

```python
df = df.drop(["name_flag", "dataset", "bnn_layer", "epochs", "activation", "final_layer_noise"], axis=1)
df
```

```python
df.to_csv("DVIP_UCI.csv", index = False)
```
