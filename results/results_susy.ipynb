```python
import os
import glob
import pandas as pd
os.chdir(".")
```

```python
res = [i for i in glob.glob('DVIP_SUSY/*')]
df = []
for f in res:
    try: 
        df.append(pd.read_csv(f))
    except:
        continue
df = pd.concat(df)
df = df.loc[:, (df != df.iloc[0]).any()] 
df = df.drop(["name_flag", "dataset", "activation", "final_layer_noise"], axis=1)
df.drop_duplicates(inplace = True)
df["ACC"] = 1 - df["Error"]
df.to_csv("DVIP_SUSY.csv", index = False)
```

```python
res2 = [i for i in glob.glob('DGP_SUSY/*')]
df2 = pd.concat([pd.read_csv(f) for f in res2 ]).drop(["Unnamed: 0"], axis = 1)
df2.to_csv("DGP_SUSY.csv", index = False)
```
