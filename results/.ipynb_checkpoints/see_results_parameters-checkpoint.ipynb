```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.patches as Circle

import numpy as np
```

Load CSV and drop columns where all the values coincide. For example, the activation used.

```python
df = pd.read_csv("results.csv")
df = df.loc[:, (df != df.iloc[0]).any()] 
```

Group table by their common parameters and compute mean and std over split

```python
def std(x):
    return np.std(x)/np.sqrt(len(x))
full_df = df.groupby(["dataset_name", "bnn_structure", "vip_layers", ]).agg(["mean", std]).drop("split", axis = 1)
full_df
```

# Relplots

```python
df = df.groupby(["dataset_name", "bnn_structure", "dropout", "epochs", "vip_layers"]).mean()
```

# Comparative with other models

```python
def comparative_plot(model_names, mean_values, std_values, colors, ax):
    """
    Creates a plot where the different models are shown in the "y" axis and mean and std values on the x axis.
    The first model is used to create a vertical dashed line, as a benchmark.
    """
    ax.scatter(mean_values, model_names, c = colors)
    ax.hlines(model_names, mean_values - std_values, mean_values + std_values, colors = colors)
    ax.axvline(mean_values[0], linestyle = "dashed", color = "k")
```

Select the configuration of the results to show.

```python
dropout = 0.1
epochs = 30000
bnn_structure = "[10, 10]"
vip_layers = ["[1]", "[2]", "[3]", "[4]", "[5]"]
```

```python
dvip_df = full_df.query("dropout == @dropout and epochs == @epochs and bnn_structure == @bnn_structure and vip_layers in @vip_layers")
```

Get results from the experiment with bnn 200-200.

```python
vip_200 =  full_df.query("dropout == @dropout and epochs == @epochs and bnn_structure == '[200, 200]' and vip_layers in @vip_layers")
```

Get Doubly's results

```python
doubly = pd.read_csv("doubly_30000.csv")
doubly = doubly.groupby(["dataset", "layers"]).agg(["mean", std]).drop("split", axis = 1)
doubly
```

Define models array with concatenated means and stds:

```python
def get_dataset(dataset, metric, agg):
    return dict(zip(model_names, np.concatenate([
        dvip_df.query("dataset_name == @dataset").xs((metric, agg),axis=1).to_numpy(),
        vip_200.query("dataset_name == @dataset").xs((metric, agg),axis=1).to_numpy(),
        doubly.query("dataset == @dataset").xs((metric, agg),axis=1).to_numpy()
])))

def plot_datasets(dataset_names, model_names, colors, metric):
    fig, ax = plt.subplots(2, len(dataset_names)//2, figsize = (20, 10))
    for i in range(len(dataset_names)):
        dataset = dataset_names[i]
        ax[i// (len(dataset_names)//2), i%(len(dataset_names)//2)].set_title(dataset.capitalize())
        ax[i// (len(dataset_names)//2), i%(len(dataset_names)//2)].locator_params(axis="x", nbins=4)
        comparative_plot(model_names, 
                         np.array([get_dataset(dataset, metric, "mean")[model] for model in model_names]), 
                         np.array([get_dataset(dataset, metric, "std")[model] for model in model_names]), 
                         [colors[model] for model in model_names], 
                         ax[i// (len(dataset_names)//2), i%(len(dataset_names)//2)])
    
        break
    handles = [
        Circle.Patch(color='darkblue', label='Single layer models'),
        Circle.Patch(color='teal', label='This work'),
        Circle.Patch(color='darkorange', label='Deep GP models'),
        Circle.Patch(color='orchid', label='Bayesian NN'),
    ]


    ax[1][1].legend(loc='upper center', bbox_to_anchor=(1.1, -0.15),
              fancybox=True, shadow=True, ncol=5, handles = handles)
```

```python
dataset_names = ["boston", "energy", "concrete", "power", "naval", "protein", "yatch", "winered"]
model_names = ["VIP", "DVIP 2", "DVIP 3", "DVIP 4", "DVIP 5", "VIP 200", "DGP", "DGP 2", "DGP 3", "DGP 4", "DGP 5"]
colors = {"VIP": "darkblue", 
          "DVIP 2": "teal", 
          "DVIP 3": "teal", 
          "DVIP 4": "teal", 
          "DVIP 5": "teal", 
          "VIP 200": "darkblue", 
          "DGP": "darkblue",
          "DGP 2": "darkorange", 
          "DGP 3": "darkorange",
          "DGP 4": "darkorange",
          "DGP 5": "darkorange"}
```

```python
plot_datasets(dataset_names, model_names, colors, "RMSE")
```

```python
plot_datasets(dataset_names, model_names, colors, "NLL")
```
