```python
import os
import glob
import pandas as pd
os.chdir("..\..\..\Desktop")
```

# Simpler Bayes Linear

```python
res1 = [i for i in glob.glob('results2/*')]
df = []
for f in res1:
    try: 
        df.append(pd.read_csv(f))
    except:
        continue
df = pd.concat(df)
df = df.loc[:, (df != df.iloc[0]).any()] 
```

```python
df = df.drop(["name_flag", "dataset", "activation", "likelihood", "final_layer_noise"], axis=1)
```

```python
df.groupby(["dataset_name", "epochs", "regression_coeffs", "vip_layers"]).mean()
```
